```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input, LSTM, GRU
from tensorflow.keras.optimizers import Adam, RMSprop, Nadam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Load dataset
file_path = '/content/electromagnetic_nano_network_traffic.csv'
data_rnn = pd.read_csv(file_path)

# Encode the categorical column 'Protocol_Type'
label_encoder = LabelEncoder()
data_rnn['Protocol_Type'] = label_encoder.fit_transform(data_rnn['Protocol_Type'])

# Split features and target after encoding
X_rnn = data_rnn.drop(columns=['Traffic_Type']).values  # Features
y_rnn = data_rnn['Traffic_Type'].values   # Target

# Split the data into training and testing sets (with stratification)
X_train_rnn, X_test_rnn, y_train_rnn, y_test_rnn = train_test_split(
    X_rnn, y_rnn, test_size=0.2, random_state=42, stratify=y_rnn
)

# Scale the data
scaler_rnn = StandardScaler()
X_train_rnn = scaler_rnn.fit_transform(X_train_rnn)
X_test_rnn = scaler_rnn.transform(X_test_rnn)

# Reshape data for RNN input (add an additional dimension for the time step, e.g., (samples, time_steps, features))
X_train_rnn = X_train_rnn.reshape((X_train_rnn.shape[0], 1, X_train_rnn.shape[1]))  # (samples, time_steps, features)
X_test_rnn = X_test_rnn.reshape((X_test_rnn.shape[0], 1, X_test_rnn.shape[1]))  # (samples, time_steps, features)

# Define the RNN model (using LSTM or GRU)
def build_rnn(model_type='LSTM'):
    model_rnn = Sequential([
        Input(shape=(X_train_rnn.shape[1], X_train_rnn.shape[2])),
        LSTM(64, activation='relu', return_sequences=False) if model_type == 'LSTM' else GRU(64, activation='relu', return_sequences=False),
        Dropout(0.4),
        Dense(32, activation='relu'),
        Dropout(0.4),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')  # Binary classification
    ])
    return model_rnn

# Custom callback for calculating precision, recall, and F1 score after each epoch
class MetricsCallback(tf.keras.callbacks.Callback):
    def __init__(self, X_val, y_val):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val
        self.precision = []
        self.recall = []
        self.f1 = []

    def on_epoch_end(self, epoch, logs=None):
        y_pred = (self.model.predict(self.X_val) > 0.5).astype(int)
        precision = precision_score(self.y_val, y_pred)
        recall = recall_score(self.y_val, y_pred)
        f1 = f1_score(self.y_val, y_pred)

        self.precision.append(precision)
        self.recall.append(recall)
        self.f1.append(f1)

# Learning rate scheduler callback
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1
)

# Dictionary to store histories of metrics for each optimizer
metric_histories = {}
training_histories = {}
final_metrics = {
    'Optimizer': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': []
}

# List of optimizers with manually specified learning rates
optimizers = {
    'Adam': Adam(learning_rate=0.001),
    'RMSprop': RMSprop(learning_rate=0.001),
    'Nadam': Nadam(learning_rate=0.001)
}

# Train the model and record metrics for each optimizer
for opt_name, opt in optimizers.items():
    # Build and compile the model
    rnn_model = build_rnn(model_type='LSTM')  # You can switch to 'GRU' here
    rnn_model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

    # Create an instance of the custom metrics callback
    metrics_callback = MetricsCallback(X_test_rnn, y_test_rnn)

    # Train the model
    history_rnn = rnn_model.fit(
        X_train_rnn, y_train_rnn,
        validation_split=0.2,
        epochs=50,
        batch_size=32,
        verbose=1,
        callbacks=[metrics_callback, lr_scheduler]
    )

    # Store the metric and training history
    metric_histories[opt_name] = metrics_callback
    training_histories[opt_name] = history_rnn

    # Evaluate on the test set
    y_pred_rnn = (rnn_model.predict(X_test_rnn) > 0.5).astype(int)
    accuracy_rnn = np.mean(y_pred_rnn.flatten() == y_test_rnn)
    precision_rnn = precision_score(y_test_rnn, y_pred_rnn)
    recall_rnn = recall_score(y_test_rnn, y_pred_rnn)
    f1_rnn = f1_score(y_test_rnn, y_pred_rnn)

    # Store the final metrics
    final_metrics['Optimizer'].append(opt_name)
    final_metrics['Accuracy'].append(accuracy_rnn)
    final_metrics['Precision'].append(precision_rnn)
    final_metrics['Recall'].append(recall_rnn)
    final_metrics['F1 Score'].append(f1_rnn)

    # Increase font sizes for plots
    plt.rcParams.update({'font.size': 18})

    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test_rnn, y_pred_rnn)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='YlOrBr')
    plt.title(f'Confusion Matrix for RNN ({opt_name})')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Display final metrics for all optimizers
metrics_df = pd.DataFrame(final_metrics)
metrics_df['Accuracy'] = metrics_df['Accuracy'].apply(lambda x: f"{x:.6f}")
print(metrics_df)

# Plot validation accuracy vs epoch
plt.figure(figsize=(12, 6))
for opt_name, history_rnn in training_histories.items():
    plt.plot(history_rnn.history['val_accuracy'], label=f'{opt_name}')
plt.title('Validation Accuracy vs Epochs for RNN Model')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid()
plt.show()

# Plot validation loss vs epoch
plt.figure(figsize=(12, 6))
for opt_name, history_rnn in training_histories.items():
    plt.plot(history_rnn.history['val_loss'], label=f'{opt_name}')
plt.title('Validation Loss vs Epochs for RNN Model')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid()
plt.show()

# Plot precision vs epoch
plt.figure(figsize=(12, 6))
for opt_name, metrics_callback in metric_histories.items():
    plt.plot(metrics_callback.precision, label=f'{opt_name}')
plt.title('Precision vs Epochs for RNN Model')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()
plt.grid()
plt.show()

# Plot recall vs epoch
plt.figure(figsize=(12, 6))
for opt_name, metrics_callback in metric_histories.items():
    plt.plot(metrics_callback.recall, label=f'{opt_name}')
plt.title('Recall vs Epochs for RNN Model')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.legend()
plt.grid()
plt.show()

# Plot F1 score vs epoch
plt.figure(figsize=(12, 6))
for opt_name, metrics_callback in metric_histories.items():
    plt.plot(metrics_callback.f1, label=f'{opt_name}')
plt.title('F1 Score vs Epochs for RNN Model')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.legend()
plt.grid()
plt.show()

# Plot ROC curve for each optimizer
plt.figure(figsize=(12, 6))
for opt_name, metrics_callback in metric_histories.items():
    # Predict probabilities for the test set
    y_proba = training_histories[opt_name].model.predict(X_test_rnn)

    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_test_rnn, y_proba)
    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    plt.plot(fpr, tpr, label=f'{opt_name} (AUC = {roc_auc:.4f})')

# Graph details
plt.title('ROC Curve for RNN Model')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
